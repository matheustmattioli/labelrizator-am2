````python
# Importando somente o necessário
import scipy
from scipy.io import arff
import pandas as pd
import numpy as np
from collections import defaultdict
````

````python
# Cada info tem 2 campos, nome do dataset e a quantidade 
# de colunas no Y
datasetsName = [('birds', 19),
                ('yeast', 14),
                ('flags', 7),
                ('genbase', 27),
                ('emotions', 6)]

# Inicializando as variaveis
data_train = [0, 0, 0, 0, 0]
meta_train = [0, 0, 0, 0, 0]
X_train = [0, 0, 0, 0, 0]
X_test = [0, 0, 0, 0, 0]
Y_train = [0, 0, 0, 0, 0]
Y_test = [0, 0, 0, 0, 0]
data_test = [0, 0, 0, 0, 0]
meta_test = [0, 0, 0, 0, 0]
````

````python
for i in range(len(datasetsName)):
    # Carregando o treino
    # dt, mt = scipy.io.arff.loadarff(f'datasets/{datasetsName[i][0]}/{datasetsName[i][0]}-train.arff')
    data_train[i], meta_train[i] = scipy.io.arff.loadarff(f'datasets/{datasetsName[i][0]}/{datasetsName[i][0]}-train.arff')
    X_train[i] = pd.DataFrame(data_train[i])

    # Carregando o teste
    data_test[i], meta_test[i] = scipy.io.arff.loadarff(f'datasets/{datasetsName[i][0]}/{datasetsName[i][0]}-test.arff')
    X_test[i] = pd.DataFrame(data_test[i])

    X_test[i].replace([b'NO',b'YES',b'0',b'1'],[0,1,0,1], inplace=True)
    X_train[i].replace([b'NO',b'YES',b'0',b'1'],[0,1,0,1], inplace=True)

    # # Separando o Y do treino
    Y_train[i] = X_train[i].iloc[:,-datasetsName[i][1]:]
    X_train[i].drop(columns=list(Y_train[i].columns), inplace=True)

    # Separando o Y do teste
    Y_test[i] = X_test[i].iloc[:,-datasetsName[i][1]:]
    X_test[i].drop(columns=list(Y_test[i].columns), inplace=True)
````
